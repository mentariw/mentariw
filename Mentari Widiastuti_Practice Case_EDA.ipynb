# Q2. How is the point distribution of the epl team? and which team is an anomaly?

```python
sns.distplot(df['general_points'])
plt.title("Distribution of General Points")
sns.despine()
```

```python
print("Skewness: %f" % df['general_points'].skew())
print("Kurtosis: %f" % df['general_points'].kurt())
```

#### The distribution of general points of EPL team is fairly symmetrical/normal

```python
#Detecting team who has anomaly in general point
Q1, Q3 = np.percentile(df['general_points'], [25 ,75])
IQR = Q3 - Q1
lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR
```

```python
outliers_low = df[df['general_points'] < lower_bound]
outliers_low
```

```python
outliers_up = df[df['general_points']>upper_bound]
outliers_up
```

```python
#Detecting outliers from z-score
from scipy import stats
import numpy as np
z = np.abs(stats.zscore(df['general_points']))
print(z)
```

```python
threshold = 3
print(np.where(z > 3))
```

#### There was no outlier; therefore, no team was considered an anomaly in terms of general points

# Q3. Which team has the best attack?

```python
#Using attack_scored
df.loc[df.attack_scored.idxmax(), ['Team']]
```

```python
best_att = df.sort_values(by = 'attack_scored', ascending=False).head(5)
```

```python
best_att[['Team','attack_scored']]
```

```python
plt.figure(figsize=(10, 8))
splot=sns.barplot(x="Team",y="attack_scored",data=best_att, color="teal")
for p in splot.patches:
    splot.annotate(format(p.get_height(), '.1f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 9), 
                   textcoords = 'offset points')
plt.xlabel("Team", size=14)
plt.ylabel("Attack scored", size=14)
```

#### In conclusion, Manchester City has the best attack shown by the highest number of goals scored

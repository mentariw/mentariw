```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
#Importing the csv data
df = pd.read_csv('https://raw.githubusercontent.com/Syukrondzeko/Fellowship-5/main/Seasons_Stats.csv')
```

```python
df
```

# Data Preprocessing

```python
#Filter data from 2017 only
season = df[df.Year == 2017]
```

```python
season
```

```python
#Checking data types
season.dtypes
```

```python
#Dropping duplicated records
season = season.drop_duplicates()
```

```python
season
#no duplicated rows
```

```python
#Checking duplicates based on Player
season.duplicated(['Player'])
```

```python
#Drop duplicates based on player
nba = season.drop_duplicates(subset='Player')
```

```python
nba
```

```python
#Checking missing values
print(nba.isnull().sum())
```

```python
#Delete columns that have as many missing values as the entire row of data --> delete columns that contain all NaN
nba_c = nba.dropna(axis=1, how='all')
```

```python
nba_c
```

# Q1. Who is the youngest and oldest player in the NBA in 2017 for each team (Tm)?

```python
#Checking outliers in Age through visualization
sns.boxplot(x=nba_c['Age'])
```

```python
#Checking outliers in Age through calculation
Q1, Q3 = np.percentile(nba_c['Age'], [25 ,75])
IQR = Q3 - Q1
lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR

outliers = nba_c[nba_c['Age']>upper_bound]
outliers
```

```python
#Finding the oldest player in each team
nba_c.loc[nba_c.groupby('Tm')['Age'].idxmax(), ['Tm', 'Player', 'Age']]
```

```python
#Finding the youngest player in each team
nba_c.loc[nba_c.groupby('Tm')['Age'].idxmin(), ['Tm', 'Player', 'Age']]
```

# Q2. Which player has the most minutes played (MP) in each position (Pos)?

```python
#Checking outliers in MP through visualization
sns.boxplot(x=nba_c['MP'])
```

```python
#Finding player with the greatest MP in each position
nba_c.loc[nba_c.groupby('Pos')['MP'].idxmax(), ['Pos', 'Player', 'Tm', 'MP']]
```

# Q3. Which team has the highest average total rebound percentage (TRB%), assist percentage (AST%), steal percentage (STL%), and block percentage (BLK%)?

```python
#Team with the highest average total rebound percentage
trb = nba_c.groupby('Tm')['TRB%'].mean().sort_values(ascending=False).head(1)
trb
```

```python
#Team with the highest average assist percentage
ast = nba_c.groupby('Tm')['AST%'].mean().sort_values(ascending=False).head(1)
ast
```

```python
#Team with the highest average steal percentage
stl = nba_c.groupby('Tm')['STL%'].mean().sort_values(ascending=False).head(1)
stl
```

```python
#Team with the highest average block percentage
blk = nba_c.groupby('Tm')['BLK%'].mean().sort_values(ascending=False).head(1)
blk
```
